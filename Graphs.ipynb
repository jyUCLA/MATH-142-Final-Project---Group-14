```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pylab import rcParams
k=2.2
rcParams['figure.figsize'] = 5*k,3*k

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df=pd.read_csv('both.csv')
data=np.array(df.drop(columns='Language'))
languages=['English',' Chinese',' Hindi',' Spanish',' Arabic',' French',' Bengali',' Portuguese',' Russian',' Urdu']

plt.ylim(0, 2.6e9)
for i in range(10):
    plt.plot(np.arange(0,51,1),np.array(df.iloc[i,1:]),label=languages[i])

plt.title('Growth of Current Top 10 Languages Over Time', fontsize=14)
plt.xlabel('Years since 2016', fontsize=12)
plt.ylabel('Number of Speakers (in Billions)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

# Show the graph
plt.savefig('both.png', bbox_inches='tight')
plt.show()
```
